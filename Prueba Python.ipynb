## 2.Contar computadoras por marca

```python
df = pd.read_csv("computadoras.csv")

conteo_ms = df['Marca'].value_counts().reset_index()

conteo_ms.columns = ['Marca', 'Cantidad']

display(conteo_ms)
```

## 4.Computadora más cara

```python
df = pd.read_csv("computadoras.csv")

computadora_mc = df.loc[df['Precio'].idxmax()]

display(computadora_mc)
```

## 5.Agregar columna calculada

```python
df = pd.read_csv("computadoras.csv")

df['Precio por GB'] = df['Precio'] / df['Almacenamiento (GB)']

display(df)
```

# Gráficos

## 6.Gráfico de barras: Promedio de precios por marca

```python
import matplotlib.pyplot as plt

df = pd.read_csv("computadoras.csv")

promedio_precios = df.groupby('Marca')['Precio'].mean()

plt.figure(figsize=(10, 6))
promedio_precios.plot(kind='bar', rot=45, color='skyblue')

plt.title('Promedio de Precios por Marca', fontsize=16)
plt.xlabel('Marca', fontsize=12)
plt.ylabel('Precio Promedio', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 7.Gráfico de líneas: Tendencia de precios promedio por año

```python
df = pd.read_csv("computadoras.csv")

promedio_por_a = df.groupby('Año')['Precio'].mean()

plt.figure(figsize=(10, 6))
plt.plot(promedio_por_a.index, promedio_por_a.values, marker='o', linestyle='-', color='blue')

plt.title('Tendencia de Precios Promedio por Año', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Precio Promedio', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 8.Gráfico circular: Proporción de estados de computadoras

```python
df = pd.read_csv("computadoras.csv")

proporcion_computadoras = df['Estado'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(proporcion_computadoras, labels=proporcion_computadoras.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'salmon'])

plt.title('Proporción de Estados de Computadoras', fontsize=16)

plt.show()
```

## 9.Diagrama de dispersión: Relación entre precio y almacenamiento

```python
df = pd.read_csv("computadoras.csv")

plt.figure(figsize=(10, 6))
plt.scatter(df['Almacenamiento (GB)'], df['Precio'], color='blue', alpha=0.6)

plt.title('Relación entre Precio y Capacidad de Almacenamiento', fontsize=16)
plt.xlabel('Almacenamiento', fontsize=12)  
plt.ylabel('Precio', fontsize=12)
plt.grid(True)
plt.tight_layout()

plt.show()
```

## 10.Histograma: Distribución de precios de computadoras usadas

```python
df = pd.read_csv("computadoras.csv")

computadoras_usadas = df[df['Estado'] == 'Usado']

plt.figure(figsize=(10, 6))
plt.hist(computadoras_usadas['Precio'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Distribución de Precios de Computadoras Usadas', fontsize=16)
plt.xlabel('Precio', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
